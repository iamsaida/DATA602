For the study of Exploratory Data Analysis, I am using the life expectancy dataset which I have taken from kaggle. This dataset has 2938 rows and 22 columns. We run through this dataset and try to analyse the data. We will focus on population and gdp based on the life expectancy of the countries listed. We will try to figure our GDP growth with increase in population and period.

Before start of the analysis, we import all the required packages like pandas, numpy, csv, matplotlib and seaborn 

```python
import pandas as pd
import numpy as np  
import csv
import matplotlib.pyplot as plt 
import seaborn as sns 
```

Displaying the dataset we have taken

```python
df=pd.read_csv('Life Expectancy Data.csv')
df
```

Showing all the columns and it's datatype of the dataset

```python
df.info()
```

List of all the unique countries in our dataset

```python
countries =set(df['Country'])
countries
```

```python
print('\nNumber of different countries :', len(countries))
```

Statistical analysis of numeric data

```python
df.describe()
```

Showing the number of missing values in each column

```python
df.isnull().sum()
```

Based on the missing values, dropping out the rows wich has missing values

```python
df1=df.dropna()
df1.info()
```

Dropping out the unwanted columns

```python
status_df= df.Status.values
new_df = df.drop(["Status"], axis=1)
status_df
```

```python
new_df
```

Renaming the columns

```python
df.rename(columns={"Life expectancy" : "Life Expectancy", "infant deaths" : "Infant Deaths", "percentage expenditure" : "Percentage Expenditure","Total expenditure" : "Total Expenditure"}, inplace=True)
df.columns
```

Mean GDP per year

```python
avg = df1.groupby('Year')['GDP'].mean()
std = df1.groupby('Year')['GDP'].std()
fig, ax = plt.subplots()
avg.plot.bar(yerr=std, ax=ax, capsize=10, rot=2)
plt.title('Mean GDP Per Year')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.tight_layout()
plt.show()
```

Boxplot of GDP and Population based on year

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 10))
df.boxplot(column='GDP', by='Year', vert=False, showfliers=False, ax=axes[0])
df.boxplot(column='Population', by='Year', vert=False, showfliers=False, ax=axes[1])
plt.tight_layout()
plt.show()
```

Finding out the correlation among all the columns

```python
df.corr()
```

Plotting the graph showing correlation among all the columns

```python
fig, axes = plt.subplots(figsize=(15, 15)) 
sns.heatmap(data=df.corr(), annot=True, linewidths=1) 
plt.title('Correlation among the columns\n')
plt.show() 
```

Histograms showing numerical values in each column

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(12, 10))
df.plot.hist( y='Adult Mortality',bins=5, color="b", ax=axes[0][0])
df.plot.hist( y='Infant Deaths',bins=5, color="r", ax=axes[0][1])
df.plot.hist( y='Alcohol',bins=5, color="orange", ax=axes[0][2])
df.plot.hist( y='Percentage Expenditure',bins=5, color="g", ax=axes[0][3])
df.plot.hist( y='Hepatitis B',bins=5, color="indigo", ax=axes[0][4])
df.plot.hist( y='Polio',bins=5, color="black", ax=axes[1][0])
df.plot.hist( y='Total Expenditure',bins=5, color="pink", ax=axes[1][1])
df.plot.hist( y='Population',bins=5, color="violet", ax=axes[1][2])
df.plot.hist( y='Income composition of resources',bins=5, color="purple", ax=axes[1][3])
df.plot.hist( y='GDP',bins=5, color="yellow", ax=axes[1][4])
plt.show()
```

Scatter plot showing difference between different columns

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20,5))
df.plot.scatter( x='Adult Mortality', y='Infant Deaths', color='r', ax=axes[0][0], title="Adult Mortality vs. infant deaths")
df.plot.scatter( x='Population', y='Adult Mortality', color='b', ax=axes[0][1], title="Population vs. Adult Mortality")
df.plot.scatter( x='Total Expenditure', y='Percentage Expenditure', color='black', ax=axes[0][2], title="Total expenditure vs. percentage expenditure")
df.plot.scatter( x='Population', y='GDP', color='purple', ax=axes[0][3], title="Population vs. GDP")
df.plot.scatter( x='Population', y='Polio', color='yellow', ax=axes[1][0], title="Population vs. Polio")
df.plot.scatter( x='GDP', y='Adult Mortality', color='g', ax=axes[1][1], title="GDP vs. Adult Mortality")
df.plot.scatter( x='Population', y='Income composition of resources', color='indigo', ax=axes[1][2], title="Population vs. Income composition of resources")
df.plot.scatter( x='GDP', y='Income composition of resources', color='orange', ax=axes[1][3], title="GDP vs. Income composition of resources")
plt.tight_layout()
plt.show()
```

Here in this module, we have performed exploratory data analysis with health data set of different countries showing their population and gdp along with their life expectancy, adult mortality etc. We Observed the datatypes of the columns, found the missing values. Calculated the mean values of each column and plotted graphs between different columns of the dataset.We plotted the hiostogram to observe how the values are changing every year and also scatter plots to see how values are dependant on other columns. We also plotted the correlation graph which shows how values are varying depending on the other values.
