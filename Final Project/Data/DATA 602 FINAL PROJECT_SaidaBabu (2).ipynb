## Data Modeling

#### Splitting the train and test data

```python
from sklearn.model_selection import train_test_split

def generate_splits():
    y = df['Permit Category']
    X = df[[x for x in df.columns if x != 'Permit Category']]

    #X = df3
    #X = df['Status','Permit Type','Contractor State','Contractor City','Project Number','Zip Code','License #']

    return train_test_split(X, y, test_size=0.2)

X_train, X_test, y_train, y_test = generate_splits()

print(f'Training examples: {X_train.shape[0]:,}')
print(f'Test examples: {X_test.shape[0]:,}')
```

### Performing Logistic Regression on the Dataset

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer

num_vars = ['Project Number','Zip Code','License #','Address Start','Address End','Valuation',
           'Census Tract','Council District','Existing Code','Proposed Code']

num_pipeline = Pipeline([('standardize_num', StandardScaler())
                        ])

processing_pipeline = ColumnTransformer(transformers=[('proc_numeric', num_pipeline, num_vars)])
```

### Creating the processing pipeline and modeling pipeline

```python
processing_pipeline
```

```python
 modeling_pipeline = Pipeline([
        ('data_processing', processing_pipeline),
        ('logreg', LogisticRegression(penalty='none'))]
    )
```

```python
modeling_pipeline
```

```python
m=modeling_pipeline.fit(X_train, y_train)
```

```python
y_hat = m.predict(X_test)
```

### Fitting the model with train data and finding the results

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_hat))
```

#### Note: As we see the accuracy is observed to be 0.78, this signifies that whenever the new applicant gives the data, this model classifies 78 pecent of the time, if he needs be granted an e-permit or should it be further reviewed by the department.

### Using GridSearch, finding the validation score and test score of Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

modeling_pipeline = Pipeline([('data_processing',processing_pipeline),('lg',LogisticRegression(penalty='l2', solver='liblinear'))])
modeling_pipeline
```

```python
from sklearn.model_selection import GridSearchCV

params = [{'lg__C':[0.1, 1, 10,100]}]
lr_gscv= GridSearchCV(estimator=modeling_pipeline, param_grid=params, scoring='roc_auc', refit=True)
lr_gscv =lr_gscv.fit(X_train,y_train)



print(f'Validation score: {lr_gscv.best_score_:.2%}')

lr_pred = lr_gscv.predict(X_test)

print(f'Test score: {lr_gscv.score(X_test, y_test):.2%}')
```

### Using another classifer to cross validate and find the best fit model

```python
from sklearn.tree import DecisionTreeClassifier
modeling_pipeline1 = Pipeline([('data_processing',processing_pipeline),('dtree',DecisionTreeClassifier())])
modeling_pipeline1
```

### Using GridSearch, finding the validation score and test score of Decision Tree classifier

```python
params = [{
 'dtree__max_depth':[1,5,10,15,20,25,50,100]}]

dt_gscv= GridSearchCV(estimator=modeling_pipeline1, param_grid=params, scoring='roc_auc', refit=True)
dt_gscv =dt_gscv.fit(X_train,y_train)

print(f'Validation score: {dt_gscv.best_score_:.2%}')

dt_pred = dt_gscv.predict(X_test)

print(f'Test score: {dt_gscv.score(X_test, y_test):.2%}')
```

```python
print(f'Logistic Regression: {lr_gscv.score(X_test, y_test):.2%}')
print(f'Decision Tree: {dt_gscv.score(X_test, y_test):.2%}')
```

##### Note:The test score value for decision tree classifier is about 87%. This is 10% more than logistic regression score. So this is a best fit model when compared.
